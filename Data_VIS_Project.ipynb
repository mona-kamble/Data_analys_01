```python
dataset['Crop_Year'].value_counts().plot(kind='bar',color='r')
plt.savefig.dataset("Crop_Year")
```

```python
sns.countplot(x="Season",data=dataset,color='g')
plt.title("Count of Season")
plt.show()
```

```python
plt.figure(figsize=(22, 6))
sns.barplot(x='Crop', y='Production', data=dataset)
plt.xticks(rotation=90)
plt.title('Total Production (in tonnes)')
plt.show()
```

```python
#Total Production by State
state_production = dataset.groupby('State_Name')['Production'].sum().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(x='State_Name', y='Production', data=state_production, color='y')
plt.title('Total Crop Production by State')
plt.xticks(rotation=90)
plt.show()
```

```python
#Crop Production Over the Years
crop_year_production = dataset.groupby(['Crop_Year', 'Crop'])['Production'].sum().unstack()
crop_year_production.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Crop Production Over the Years')
plt.ylabel('Production')
plt.xlabel('Year')
plt.legend(title='Crop')
plt.show()
```

```python
#Area VS Production
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Area', y='Production', data=dataset, color='r')
plt.title('Area vs. Production of Crops')
plt.xlabel('Area (in hectares)')
plt.ylabel('Production (in tonnes)')
plt.show()
```
